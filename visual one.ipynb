```python
%matplotlib inline
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
sales = pd.read_csv('sales.csv', parse_dates=['Date'])
```

```python
sales
```

```python
sales['month'] = sales['Date'].dt.month
```

```python
data = sales.groupby(['month', 'Product'])['Amount'].sum()
```

```python
data
```

```python
data = data.unstack()

revenue = data.fillna(0).cumsum()
```

```python
revenue
```

```python
ax = revenue.plot.area(title='Revenue 2017 (Euro)')
```

```python
ab = revenue.plot.bar(title = 'Revenue 2017 (Euro)', grid=True)
```

```python
ac = revenue.plot.line(title = 'revenue 2017 (Euro)', grid=True)
```

```python
ad = revenue.plot.barh(title = 'revenue 2017 (Euro)', grid=True)
```

```python
total_rev = sales.groupby(['month','Product'])['Amount'].sum().cumsum()
```

```python
total_rev = total_rev.unstack()
```

```python
af = total_rev.plot.pie(subplots=True, title = 'Revenue percentage by Month')
```
